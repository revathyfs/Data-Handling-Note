# **Pandas**


---

**Pandas** is a **Python library** that helps you work with **tables of data** easily.
It lets you **store, clean, explore, and analyze data**â€”just like you would in **Excel**, but with **more power and flexibility**.

Think of Pandas as your **data toolbox**:

* **DataFrame** â†’ like a full table with rows and columns.
* **Series** â†’ like a single column from that table.

With Pandas, you can:

* **Read** data from Excel, CSV, or databases.
* **Clean** messy data (remove duplicates, fill missing values, etc.).
* **Select & filter** the parts you need.
* **Analyze & summarize** data quickly.

---

ðŸ’¡ **Simple analogy**:
If your data is a messy pile of papers, **Pandas** is the organized filing cabinet + calculator that makes everything neat and easy to work with.

---

**Built-in methods** 

You can use to work with your data.
They help you **read**, **explore**, **clean**, **analyze**, and **save** your data.

---

### **1ï¸âƒ£ Reading & Saving Data**

| Function          | Purpose                   | Example                      |
| ----------------- | ------------------------- | ---------------------------- |
| `pd.read_csv()`   | Read data from a CSV file | `pd.read_csv('data.csv')`    |
| `pd.read_excel()` | Read data from Excel      | `pd.read_excel('file.xlsx')` |
| `to_csv()`        | Save DataFrame to CSV     | `df.to_csv('output.csv')`    |
| `to_excel()`      | Save DataFrame to Excel   | `df.to_excel('output.xlsx')` |

---

### **2ï¸âƒ£ Checking Data**

| Function     | Purpose                  | Example         |
| ------------ | ------------------------ | --------------- |
| `head()`     | First 5 rows             | `df.head()`     |
| `tail()`     | Last 5 rows              | `df.tail()`     |
| `info()`     | Column info & data types | `df.info()`     |
| `describe()` | Summary statistics       | `df.describe()` |
| `shape`      | Rows & columns count     | `df.shape`      |

---

### **3ï¸âƒ£ Selecting & Filtering**

| Function  | Purpose                | Example                |
| --------- | ---------------------- | ---------------------- |
| `loc[]`   | Select by labels       | `df.loc[0, 'Name']`    |
| `iloc[]`  | Select by positions    | `df.iloc[0, 1]`        |
| `[]`      | Select column(s)       | `df['Age']`            |
| `query()` | Filter with conditions | `df.query('Age > 25')` |

---

### **4ï¸âƒ£ Cleaning Data**

| Function            | Purpose               | Example                            |
| ------------------- | --------------------- | ---------------------------------- |
| `dropna()`          | Remove missing rows   | `df.dropna()`                      |
| `fillna()`          | Fill missing values   | `df.fillna(0)`                     |
| `drop_duplicates()` | Remove duplicate rows | `df.drop_duplicates()`             |
| `rename()`          | Rename columns        | `df.rename(columns={'Old':'New'})` |

---

### **5ï¸âƒ£ Analyzing Data**

| Function         | Purpose             | Example                                  |
| ---------------- | ------------------- | ---------------------------------------- |
| `sum()`          | Sum values          | `df['Sales'].sum()`                      |
| `mean()`         | Average             | `df['Price'].mean()`                     |
| `value_counts()` | Count unique values | `df['City'].value_counts()`              |
| `groupby()`      | Group & summarize   | `df.groupby('Category')['Sales'].mean()` |
| `sort_values()`  | Sort by column      | `df.sort_values('Age')`                  |

### **Definition**

`df.to_string()` is a Pandas method that **converts the entire DataFrame into a string format** so it can be printed exactly as it looks in table form.

By default, when you just do:

```python
print(df)
```

Pandas might **truncate** (cut off) rows or columns if the DataFrame is large.

But:

```python
print(df.to_string())
```

will **print the whole DataFrame without cutting anything**.

---

### **Key Points**

* **Purpose**: To display **all** rows & columns without truncation.
* **Common Use**: When printing DataFrames for debugging or reporting.
* **Extra Options**:

  * `df.to_string(index=False)` â†’ hides row numbers.
  * `df.to_string(columns=['Name', 'Age'])` â†’ prints only selected columns.

---

ðŸ’¡ **Analogy**
Normally, Pandas shows you a **preview** of your table (like a movie trailer).
`df.to_string()` says: *"No previews, show me the **entire movie**!"*

#### Column Removal Functions in Pandas


| Function                | Description                                                                                                                                   | Use Case                                                                                |
| ----------------------- | --------------------------------------------------------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------- |
| **`pop()`**             | Removes a **single column** from the DataFrame and returns it as a **Series**. Works in place (original DataFrame changes).                   | When you need to delete a column **and** store its values for later use.                |
| **`drop()`**            | Removes rows or columns by label. Returns a **new DataFrame** unless `inplace=True`. Can remove multiple columns. Needs `axis=1` for columns. | When you want to remove rows or columns while keeping the original DataFrame unchanged. |
| **`del` keyword**       | Deletes a **single column** in place. Does **not** return it.                                                                                 | When you need to quickly delete one column and donâ€™t need its data afterward.           |
| **`drop(columns=...)`** | Shortcut for removing **multiple columns** without specifying `axis=1`. Returns new DataFrame unless `inplace=True`.                          | When you want to remove multiple columns with cleaner, more readable code.              |

| -------------------------------------- | --------------------------------------------------------------------------------------------- | ---------------------------------------------------------- |

| **`columns`**                          | Returns the **column labels** as an Index object.                                             | Check column names.                                        |

#### Data Type Inspection Functions in Pandas

| Function                                | Description                                                                                     | Use Case                                                   |

| **`dtype`** *(Series attribute)*        | Returns the **data type** of a single pandas Series.                                            | Check the type of one column.                              |

| **`type()`** *(Python built-in)*        | Returns the **Python type** of the given object.                                                | Check whether an object is a DataFrame, Series, list, etc. |

| **`infer_objects()`**                   | Attempts to infer better dtypes for object columns.                                             | Upgrade object columns to more specific types.             |
